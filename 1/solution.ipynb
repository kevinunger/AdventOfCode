```python
number_strings = [
  "one", 
  "two", 
  "three",
  "four",
  "five",
  "six",
  "seven",
  "eight",
  "nine"
]
```

```python
test_words = [
  "51591twosix4dhsxvgghxq", # 5 4
  "425nine",# 4 5
  "llvmhjtr8nbbhrfone", # 8 8,
  "lpbjvpbtdfvtxtdvkpjs7qrvddkzmjtlqtg",
  "3sixnineseven",
  "rfmsqbkms7three",
  "gbseven9five6",
  "nine2hdltdjdp73phzrjnonegx",
  "3two3eightjszbfourkxbh5twonepr",
  "5cfprzgxtf3465five",
  "eightwothree",
  "374",
  "4pshnstwo46fivetwoone",
  "5jlb4",
  "36foursixsevencjffhxrsq",
  "sevenseventwo9xthreevzfdmfqvdf6"
]

test_solutions = [
  54,
  49,
  81,
  77,
  37,
  73,
  76,
  91,
  32,
  55,
  83,
  34,
  41,
  54,
  37,
  76
]
```

```python
"""
replace all number_string with numbers
"""
def replace_number_strings(word: str) -> str:
  _translated_word = word 
  for i in range(len(word)):
    translated_word = _translated_word[:i+1]
    #(translated_word)
    for idx, number_string in enumerate(number_strings):
      number = idx + 1
      number_string_index = translated_word.find(number_string)
      if(number_string_index != -1):
        _translated_word = _translated_word.replace(number_string, str(number))
        # print("--> replaced ", number_string, " ", _translated_word)

  return _translated_word

print(replace_number_strings("eight2testtestfourfive"))
```

```python
def get_numbers(word: str) -> int:
  word = replace_number_strings(word)
  first_number = 0
  last_number = 0
  for letter in word:
    if (letter.isnumeric()):
      if not first_number:
        first_number = letter

      if first_number != 0:
        last_number = letter
      
  number = first_number + last_number # concat strings "1" + "2" == "12"
  return int(number)
```

```python
x = get_numbers(replace_number_strings("dc8"))
print(x)
```

```python
# test
for i, test_word in enumerate(test_words):
  if (get_numbers(test_word) != test_solutions[i]):
    print("error with ", i)
    print(get_numbers(test_word), test_solutions[i])
```

```python
def get_total_sum(lines):
  sum = 0
  for i, line in enumerate(lines):
    line = line.replace("\n", "")
    line_replaced_number_strings = replace_number_strings(line)
    sum += get_numbers(line_replaced_number_strings)
    print(f"{line} -> {line_replaced_number_strings} -> {get_numbers(line_replaced_number_strings)}")

  return sum
```

```python
file = open("input.txt", "r")
lines = file.readlines()
sum = get_total_sum(lines)

print(sum)
```
